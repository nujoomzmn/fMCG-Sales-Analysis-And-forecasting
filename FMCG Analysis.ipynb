```python
import pandas as pd
from sqlalchemy import create_engine
from urllib.parse import quote_plus


server = r'NUJOOMZMN\SQLEXPRESS'
database = 'FMCG_Data'
table_name = 'FMCG_DATA' 
driver = 'ODBC Driver 17 for SQL Server'

params = quote_plus(f"DRIVER={driver};SERVER={server};DATABASE={database};Trusted_Connection=yes")

# Create the engine
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

# Insert the dataframe into SQL Server
data.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

print(f"✅ DataFrame successfully inserted into {database}.{table_name}")
```
